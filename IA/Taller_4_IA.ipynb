```python
## JUAN FELIPE GUALDRIA CARRANZA / NICOLAS IGNACIO SANCHEZ VARGAS
## Red Neuronal No Convolucional
import tensorflow as tf
from tensorflow.keras import models, layers
from tensorflow.keras.datasets import cifar10
import matplotlib.pyplot as plt
from tensorflow.keras.utils import to_categorical

# Cargar el dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Normalizar los datos
x_train = x_train.reshape((x_train.shape[0], -1)) / 255.0
x_test = x_test.reshape((x_test.shape[0], -1)) / 255.0

# Codificar las etiquetas como categorías
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Definir la red neuronal
model_fc = models.Sequential([
    layers.Dense(512, activation='relu', input_shape=(x_train.shape[1],)),
    layers.Dropout(0.5),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax')
])

# Compilar la red
model_fc.compile(optimizer='adam',
                 loss='categorical_crossentropy',
                 metrics=['accuracy'])

# Entrenar la red
history_fc = model_fc.fit(x_train, y_train, 
                          epochs=10, 
                          batch_size=64, 
                          validation_data=(x_test, y_test))
```

```python
##Red Neuronal Convolucional
# Redefinir el dataset con su estructura original y normalizar
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
x_train = x_train / 255.0
x_test = x_test / 255.0
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Definir la red convolucional
model_cnn = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compilar la red
model_cnn.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

# Entrenar la red
history_cnn = model_cnn.fit(x_train, y_train, 
                            epochs=10, 
                            batch_size=64, 
                            validation_data=(x_test, y_test))
```

```python
# Función para graficar el desempeño
def plot_history(history, title):
    plt.plot(history.history['accuracy'], label='Entrenamiento')
    plt.plot(history.history['val_accuracy'], label='Validación')
    plt.title(title)
    plt.xlabel('Épocas')
    plt.ylabel('Precisión')
    plt.legend()
    plt.show()

# Graficar desempeño de la red densa
plot_history(history_fc, "Desempeño de la Red No Convolucional")

# Graficar desempeño de la red convolucional
plot_history(history_cnn, "Desempeño de la Red Convolucional")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Aumento de datos
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True
)

datagen.fit(x_train)

# Red con Dropout y regularización L2
from tensorflow.keras.regularizers import l2

model_cnn_improved = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', kernel_regularizer=l2(0.001), input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.3),
    layers.Conv2D(64, (3, 3), activation='relu', kernel_regularizer=l2(0.001)),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.3),
    layers.Conv2D(64, (3, 3), activation='relu', kernel_regularizer=l2(0.001)),
    layers.Flatten(),
    layers.Dense(64, activation='relu', kernel_regularizer=l2(0.001)),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax')
])

model_cnn_improved.compile(optimizer='adam',
                           loss='categorical_crossentropy',
                           metrics=['accuracy'])

# Entrenamiento con aumento de datos
history_cnn_improved = model_cnn_improved.fit(
    datagen.flow(x_train, y_train, batch_size=64),
    epochs=20,
    validation_data=(x_test, y_test)
)

# Graficar el desempeño de la red mejorada
plot_history(history_cnn_improved, "Desempeño de la Red Mejorada")
```

```python
def plot_comparison(histories, labels, title):
    plt.figure(figsize=(10, 6))
    
    for history, label in zip(histories, labels):
        plt.plot(history.history['val_accuracy'], label=f'Validación - {label}')
        plt.plot(history.history['accuracy'], '--', label=f'Entrenamiento - {label}')
    
    plt.title(title)
    plt.xlabel('Épocas')
    plt.ylabel('Precisión')
    plt.legend()
    plt.grid()
    plt.show()

# Comparar las redes
plot_comparison(
    [history_fc, history_cnn, history_cnn_improved],
    ["No Convolucional", "Convolucional", "Convolucional Mejorada"],
    "Comparación del Desempeño"
)
```
